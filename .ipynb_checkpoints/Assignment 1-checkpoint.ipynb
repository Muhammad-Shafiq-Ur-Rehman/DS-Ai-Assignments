## Task 7

```python
phy= float(input("Enter marks for Physics :"))
chem= float(input("Enter marks for Chemistry :"))
math= float(input("Enter marks for Math :"))
eng= float(input("Enter marks for English :"))
urdu= float(input("Enter marks for Urdu :"))
total = phy+chem+math+eng+urdu
percentage = (total/500)*100
if (percentage>100) or (percentage<0):
    print("Error in calculating percentage")
elif percentage >=90 and percentage <=100:
    print("A+ grade")
elif percentage >=80 and percentage <90:
    print("A grade")
elif percentage >=70 and percentage <80:
    print("B grade")
elif percentage >=60 and percentage <70:
    print("C grade")
elif percentage >=50 and percentage <60:
    print("D grade")
elif percentage >=40 and percentage <50:
    print("E grade")
else:
    print("F grade")
    
```
